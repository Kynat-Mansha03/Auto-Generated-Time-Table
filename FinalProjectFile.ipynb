Reading Entities from the file and storing it in lists one for each column

```python
import pandas as pd
import random 
# Read the data using pandas
df = pd.read_excel('Teachers.xlsx')

# Display the DataFrame


# Convert each column to a list and remove duplicates
section = list(set(df['Section'].tolist()))
course_instructor = list(set(df['Course Instructor'].tolist()))
course = list(set(df['Course'].tolist()))
df['Students'] = df['Students'].astype(str)
students = list(set(df['Students'].tolist()))
#assign 7 bits 
#start from 0
# 0 0 0 0 0 0 0
sectionbits = [bin(i)[2:].zfill(7) for i in range(len(section))]
course_instructorbits = [bin(i)[2:].zfill(7) for i in range(len(course_instructor))]
coursebits = [bin(i)[2:].zfill(7) for i in range(len(course))]
studentsbits = [bin(i)[2:].zfill(7) for i in range(len(students))]

print(studentsbits)
 #print(course_instructor)
#print(course)
print(coursebits)
```

Binary encoding

```python
# import pandas as pd
# import random 
data=pd.read_excel('Teachers.xlsx')

Teachers_info = []

for _, row in data.iterrows():
    # Create binary encoded strings for each column
    course_binary = str(row['Course'])   
    #look for in course and with the same index replace the value with the value in coursebits for same index
    course_binary = coursebits[course.index(course_binary)]
    theory_lab_binary = '0000001' 
    section_binary = str(row['Section']) 
    section_binary = sectionbits[section.index(section_binary)]
    students_binary = str(row['Students']) 
    students_binary = studentsbits[students.index(students_binary)]
    professor_binary = str(row['Course Instructor']) 
    professor_binary = course_instructorbits[course_instructor.index(professor_binary)]

    # Concatenate the binary encoded strings into a single string
    row_info = ''.join([course_binary, theory_lab_binary, section_binary, students_binary, professor_binary])

    # Append the binary encoded row to the list
    Teachers_info.append(row_info)
    row_info=''
#print(section_binary)
for i in range(12):
    index = random.randint(0, len(Teachers_info) - 1)
    Teachers_info[index] = Teachers_info[index][:7] + '0000000' + Teachers_info[index][14:]

    
print(Teachers_info)
 
```

Initializing Time Slots and Room Numbers 

```python
Time_slots_theory=['8.30-9.50','10.05-11.25','11.40-1','1.15-2.35','2.50-4.10','4.25-5.55','6.10-7.40']
Rooms =['C-401', 'C-402', 'C-403', 'C-404', 'C-405', 'C-406', 'C-407', 'C-408', 'C-409', 'C-410', 'B-227', 'B-229', 'B-230']
Halls=['C-301', 'C-302', 'C-303', 'C-304', 'C-305', 'C-306']
Days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday']

#convert to 7 bits
#start from 0
# 0 0 0 0 0 0 0
Time_slots_theorybits = [bin(i)[2:].zfill(7) for i in range(len(Time_slots_theory))]
Hallsbits = [bin(i)[2:].zfill(7) for i in range(len(Halls))]
Roomsbits = [bin(i)[2:].zfill(7) for i in range(len(Rooms))]
Daysbits = [bin(i)[2:].zfill(7) for i in range(len(Days))]

#assigning all the rooms and the halls against the time slots 
#these halls and rooms get assigned, they get removed from the list 
schedule = {time_slot: {'Rooms': Roomsbits, 'Halls': Hallsbits} for time_slot in Time_slots_theorybits}

print(Hallsbits)

 
```

```python
#Course, Theory/Lab, Section, Section-Strength, Professor, First-lecture-day, 
# First- lecture-timeslot, First-lecture-room, First-lecture-room-size, Second
# lecture-day, Second-lecture-timeslot, Second-lecture-room, Second-lecture
# room-size 
def decoder(chromosome):
    # Unpack the chromosome string
    info = chromosome['info']
    first_lecture_day_bits = chromosome['First Lecture Day']
    first_lecture_timeslot_bits = chromosome['First Lecture Timeslot']
    first_lecture_room_bits = chromosome['First Lecture Room']
    first_lecture_room_size_bits = chromosome['First Lecture Room Size']
    second_lecture_day_bits = chromosome['Second Lecture Day']
    second_lecture_timeslot_bits = chromosome['Second Lecture Timeslot']
    second_lecture_room_bits = chromosome['Second Lecture Room']
    second_lecture_room_size_bits = chromosome['Second Lecture Room Size']

    # Decode the info part
    course_info_bits = info[0:7]
    theory_lab_bits = info[7:14]
    section_info_bits = info[14:21]
    professor_info_bits = info[28:35]
    student_size_bits=info[21:28]

     
    course_index = coursebits.index(course_info_bits)
    course_name = course[course_index]
     

    # Decode theory_lab
    theory_lab_type = 'Theory' if theory_lab_bits == '0000001' else 'Lab'

    # Decode section
    section_index = sectionbits.index(section_info_bits)
    section_name = section[section_index]

    # Decode students
    # students_count = studentsbits.index(first_lecture_room_size_bits)
    # students_count = students[students_count]

    # Decode professor
    professor_index = course_instructorbits.index(professor_info_bits)
    professor_name = course_instructor[professor_index]
    #Decoding student size
    student_index=studentsbits.index(student_size_bits)
    student_size=students[student_index]
    # Decode first lecture details
    first_lecture_day_index = Daysbits.index(first_lecture_day_bits)
    first_lecture_day = Days[first_lecture_day_index]
    first_lecture_timeslot_index = Time_slots_theorybits.index(first_lecture_timeslot_bits)
    first_lecture_timeslot = Time_slots_theory[first_lecture_timeslot_index]
    first_lecture_room_size = student_size
    #first_lecture_room = Rooms[Roomsbits.index(first_lecture_room_bits)] if 60 < int(students_count) <= 120 else Halls[Hallsbits.index(first_lecture_room_bits)]
    if int(student_size)>60 and int(student_size)<=120:
        first_lecture_room = Halls[Hallsbits.index(first_lecture_room_bits)]
    else:
         first_lecture_room = Rooms[Roomsbits.index(first_lecture_room_bits)]
        
    # Decode second lecture details
    second_lecture_day_index = Daysbits.index(second_lecture_day_bits)
    second_lecture_day = Days[second_lecture_day_index]
    second_lecture_timeslot_index = Time_slots_theorybits.index(second_lecture_timeslot_bits)
    second_lecture_timeslot = Time_slots_theory[second_lecture_timeslot_index]
    second_lecture_room_size = student_size
    #second_lecture_room = Rooms[Roomsbits.index(second_lecture_room_bits)] if 60 < int(students_count) <= 120 else Halls[Hallsbits.index(second_lecture_room_bits)]
    if int(student_size)>60 and int(student_size)<=120:
        second_lecture_room = Halls[Hallsbits.index(second_lecture_room_bits)]
    else:
         second_lecture_room = Rooms[Roomsbits.index(second_lecture_room_bits)]
    # Return the decoded information
    return {
        'Course': course_name,
        'Theory/Lab': theory_lab_type,
        'Section': section_name,
        'Students': student_size,
        'Professor': professor_name,
        'First Lecture Day': first_lecture_day,
        'First Lecture Timeslot': first_lecture_timeslot,
        'First Lecture Room': first_lecture_room,
        'First Lecture Room Size': student_size,
        'Second Lecture Day': second_lecture_day,
        'Second Lecture Timeslot': second_lecture_timeslot,
        'Second Lecture Room': second_lecture_room,
        'Second Lecture Room Size': student_size
    }
```

```python

 

# def CrossOver(individual1, individual2):
#     child1 = Timetable()
#     child2 = Timetable()

#     # Get random indices for crossover
#     crossOverIndex1 = random.randint(10, len(individual1.chromosomes))
#     crossOverIndex2 = random.randint(crossOverIndex1, len(individual1.chromosomes))

#     # Perform multiple crossover
#     child1.chromosomes = individual1.chromosomes[:crossOverIndex1] + individual2.chromosomes[crossOverIndex1:crossOverIndex2] + individual1.chromosomes[crossOverIndex2:]
#     child2.chromosomes = individual2.chromosomes[:crossOverIndex1] + individual1.chromosomes[crossOverIndex1:crossOverIndex2] + individual2.chromosomes[crossOverIndex2:]

#     return child1, child2

# def CrossOver(individual1, individual2):
#     child1 = Timetable()
#     child2 = Timetable()

#     # Define the length of chromosomes
#     chromo_length = len(individual1.chromosomes)

#     # Initialize crossover masks for both children
#     mask1 = [random.choice([True, False]) for _ in range(chromo_length)]
#     mask2 = [not m for m in mask1]

#     # Perform crossover
#     child1.chromosomes = [chromo1 if m1 else chromo2 for chromo1, chromo2, m1 in zip(individual1.chromosomes, individual2.chromosomes, mask1)]
#     child2.chromosomes = [chromo1 if m2 else chromo2 for chromo1, chromo2, m2 in zip(individual1.chromosomes, individual2.chromosomes, mask2)]

#     return child1, child2


def CrossOver(individual1, individual2):
    child1 = Timetable()
    child2 = Timetable()

    # Get the length of the chromosome
    ARRAY_SIZE = len(individual1.chromosomes)

    # Get random indices for crossover
    crossover_point_p1_01 = random.randint(1, ARRAY_SIZE - 1)
    crossover_point_p2_01 = random.randint(1, ARRAY_SIZE - 1)

    # Initialize child chromosomes
    child1.chromosomes = individual1.chromosomes[:crossover_point_p1_01] + [0] * (ARRAY_SIZE - crossover_point_p1_01)
    child2.chromosomes = individual2.chromosomes[:crossover_point_p2_01] + [0] * (ARRAY_SIZE - crossover_point_p2_01)

    # Perform crossover
    for i in range(ARRAY_SIZE):
        if individual2.chromosomes[(crossover_point_p1_01 + i) % ARRAY_SIZE] in child1.chromosomes:
            index = child1.chromosomes.index(individual2.chromosomes[(crossover_point_p1_01 + i) % ARRAY_SIZE])
            child1.chromosomes[index], child1.chromosomes[(crossover_point_p1_01 + i) % ARRAY_SIZE] = child1.chromosomes[(crossover_point_p1_01 + i) % ARRAY_SIZE], child1.chromosomes[index]
        else:
            child1.chromosomes[(crossover_point_p1_01 + i) % ARRAY_SIZE] = individual2.chromosomes[(crossover_point_p1_01 + i) % ARRAY_SIZE]

    for i in range(ARRAY_SIZE):
        if individual1.chromosomes[(crossover_point_p2_01 + i) % ARRAY_SIZE] in child2.chromosomes:
            index = child2.chromosomes.index(individual1.chromosomes[(crossover_point_p2_01 + i) % ARRAY_SIZE])
            child2.chromosomes[index], child2.chromosomes[(crossover_point_p2_01 + i) % ARRAY_SIZE] = child2.chromosomes[(crossover_point_p2_01 + i) % ARRAY_SIZE], child2.chromosomes[index]
        else:
            child2.chromosomes[(crossover_point_p2_01 + i) % ARRAY_SIZE] = individual1.chromosomes[(crossover_point_p2_01 + i) % ARRAY_SIZE]

    return child1, child2
 

def tournament_selection(population, fitness_scores, tournament_size):
    tournament = random.sample(list(zip(population, fitness_scores)), tournament_size)
    return max(tournament, key=lambda x: x[1])[0]

def genetic_algorithm():
    population = generatePopulation()
    best_individuals = []
    best_fitness_score = float('inf')

    for generation in range(GENERATION):
        fitness_scores = [EvaluateFitness(individual) for individual in population]
        # print(fitness_scores)
        # print('\n')
        best_fitness_score = max(fitness_scores)
        best_individuals = [individual for individual, fitness in zip(population, fitness_scores) if fitness == best_fitness_score]

        if best_fitness_score == 0:
            break

        #Select parents using tournament selection
        parent1 = tournament_selection(population, fitness_scores, 3)
        parent2 = tournament_selection(population, fitness_scores, 3)
        while parent1 == parent2:
            parent2 = tournament_selection(population, fitness_scores, 3)
        #population = sorted(population, key=EvaluateFitness)
    
        #newpopulation=[]
        #for i in range(0, 25):
            # child1, child2 = makeCrossSingleOver(population[i], population[i+1], len(population[i].chromosomes) // 2)
            # newpopulation.append(child1)
            # newpopulation.append(child2)
        #child1, child2 = makeCrossSingleOver(parent1, parent2, len(parent1.chromosomes) // 2)
            
        # for i in range(25, 50):
        #     child1, child2 = makeMultiCrossOver(population[i], population[i+1], len(population[i].chromosomes) // 3, 2 * len(population[i].chromosomes) // 3)
        #     newpopulation.append(child1)
        #     newpopulation.append(child2)
       # child1, child2 = makeMultiCrossOver(parent2, parent2, len(parent1.chromosomes) // 3, 2 * len(parent1.chromosomes) // 3)
        # for i in range(50, 99):
        #     child1, child2 = makeRandomCrossOver(population[i], population[i+1])
        #     newpopulation.append(child1)
        #     newpopulation.append(child2)
       # child1, child2 = makeRandomCrossOver(parent1, parent2)
        #newpopulation=sorted(newpopulation, key=EvaluateFitness)
        # population=newpopulation[:100]
        
        child1,child2=CrossOver(parent1,parent2)
        child1=Mutation(child1)
        child2=Mutation(child2)
        if child1 not in population:
         worst_index = fitness_scores.index(min(fitness_scores))
         population[worst_index] = child1
        if child2 not in population:
            worst_index = fitness_scores.index(min(fitness_scores))
            population[worst_index] = child2

    return best_individuals, best_fitness_score


best_individuals, best_fitness_score = genetic_algorithm()
#print(best_individuals)
for individual in best_individuals:
    for chromosome in individual.chromosomes:
        decoded_info = decoder(chromosome)
        print(decoded_info)
        print('\n')
print(best_fitness_score)
```
